Estas búsquedas son demasiado específicas en general como para sacar conclusiones

Ahora quiero ver cuántas búsquedas se hacen especificando la capacidad de memoria

```python
busquedasConMemoria = pd.DataFrame
busquedasConMemoria = {'32GB': busquedas.str.contains('32gb|32 gb|32g|32 g|32 gigas|32gigas|32bgigas|32bg|32 bg', regex = True).value_counts()\
                ,'64GB': busquedas.str.contains('64gb|64 gb|64g|64 g|64 gigas|64gigas|64bgigas|64bg|64 bg', regex = True).value_counts()\
                ,'16GB': busquedas.str.contains('16gb|16 gb|16g|16 g|16 gigas|16gigas|16bgigas|16bg|16 bg', regex = True).value_counts()\
                ,'128GB': busquedas.str.contains('128gb|128 gb|128g|128 g|128 gigas|128gigas|128bgigas|128bg|128 bg', regex = True).value_counts()}
busquedasConMemoriaDF = pd.DataFrame(data=busquedasConMemoria)
busquedasConMemoriaDF = busquedasConMemoriaDF.drop(busquedasConMemoriaDF.index[0])
busquedasConMemoriaDF = busquedasConMemoriaDF.transpose()
busquedasConMemoriaDF.columns = ['Cantidad de busquedas']
```

```python
busquedasConMemoriaDF.plot(kind = 'bar')
```

Se observa que las búsquedas de 32GB podrían ser las más comunes, sin embargo una persona puede tipear lo mismo de muchas maneras, por lo que podría haber más búsquedas referidas a la capacidad de memoria de las que se contemplaron.

```python
busquedasPorModelo = pd.DataFrame
busquedasPorModelo = {'Motorola': busquedas.str.contains('moto', regex = True).value_counts()\
                ,'iPhone': busquedas.str.contains('iphone|iphon|i phone|i phon', regex = True).value_counts()\
                ,'Asus': busquedas.str.contains('asus', regex = True).value_counts()\
                ,'Samsung': busquedas.str.contains('samsung|sansung|samsu|sansu', regex = True).value_counts()\
                ,'Blackberry': busquedas.str.contains('blackberry|black berry|bery', regex = True).value_counts()}
busquedasPorModeloDF = pd.DataFrame(data=busquedasPorModelo)
busquedasPorModeloDF = busquedasPorModeloDF.drop(busquedasPorModeloDF.index[0])
busquedasPorModeloDF = busquedasPorModeloDF.transpose()
busquedasPorModeloDF.columns = ['Cantidad de busquedas']
busquedasPorModeloDF
```

```python
busquedasPorModeloDF.plot(kind = 'bar')
```

Esto indica que el modelo de teléfono sobre el que más se hacen búsquedas es el iPhone, lo que parece coincidir con la cantidad de eventos viewed product que disponen los modelos, vistos en 'Análisis sobre modelos'

Por desgracia al igual que antes, es difícil determinar todas las formas posibles en la que una persona puede escribir mal el nombre de un modelo en una búsqueda. Por lo que este análisis no tiene garantía de exactitud
