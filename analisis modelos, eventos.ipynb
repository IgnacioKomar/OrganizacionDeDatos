```python
import pandas as pd
labels_train = pd.read_csv('labels_training_set.csv')
eventos=pd.read_csv('events_up_to_01062018.csv')
labels_test=pd.read_csv('trocafone_kaggle_test.csv')
#labels_train.head()
```

```python
eventos['timestamp']=pd.to_datetime(eventos['timestamp'])
```

```python
eventos2=eventos.copy()
#tomo los 100 modelos mas populares
modelos_populares=eventos2[eventos2['event']=='conversion'].groupby(['model'])['model'].count().rename('cuenta').sort_values(ascending=False).reset_index().head(100).model.tolist()
#type(modelos_populares)
modelos_populares
```

```python
eventos_onehot = eventos[['person','event','model']].copy()
#hago un one hot encoding de la columna event
eventos_onehot = pd.get_dummies(eventos_onehot ,columns=['event'],prefix=[''],prefix_sep='')

#filtro los datos que contengan alguno de los modelos populares 
eventos_onehot=eventos_onehot[eventos_onehot['model'].apply(lambda x: x in modelos_populares)]
#hago un one hot encoding de la columna model
eventos_onehot = pd.get_dummies(eventos_onehot ,columns=['model'],prefix='',prefix_sep='')
#hago la suma de la cantidad de eventos por persona
eventos_sum=eventos_onehot.groupby(['person']).sum().reset_index()

eventos_sum.describe()
```

```python
# preprwaro los dataframe de entrenamiento de testeo
eventos_train=labels_train.merge(eventos_sum,how='left',on='person').fillna(0)
eventos_test=labels_test.merge(eventos_sum,how='left',on='person').fillna(0)
eventos_train.describe()
```

```python
#guardo los features
#features=eventos_train.drop(['person' , 'label','checkout','conversion','lead','viewed product'] , axis='columns').columns
features2=['checkout','conversion','lead','viewed product']
#guardo los features
features=eventos_train.drop(['person' , 'label'] , axis='columns').columns
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
y = eventos_train['label']
predictor_cols = features
 

# Create training predictors data
X = eventos_train[predictor_cols]
seed = 0
test_size = 0.5
train_X, X_test, train_y, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
#model = XGBClassifier()
xgb_model=xgb.XGBClassifier(max_depth=4, n_estimators=300,eta=0.05 ,colsample_bytree=0.3,gamma=6,subsample=1,learning_rate=0.06,objective='binary:logistic')
xgb_model.fit(train_X, train_y)

scores = cross_val_score(xgb_model,train_X, train_y, scoring="roc_auc", cv=3)
scores.mean()
```

```python
from sklearn.ensemble import VotingClassifier
vc=VotingClassifier(estimators=[('xgb',xgb_model),('GBC',GBC)],voting='soft')
scores = cross_val_score(vc, train_X, train_y, scoring="roc_auc", cv=4)
print(scores.mean())
```
